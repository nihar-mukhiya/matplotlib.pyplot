# matplotlib.pyplot

### matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB.

##### in below graph the funtion plot() automatically generates x values bcoz if single array given as input it considers them as y values

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

##### in below graph we specify x and y values

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

##### axis() allows us to specify range of values on x-axis and y-axis

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.show()
```

##### different formats of displaying graph

```python

t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, 'r--', t**2, 'bs', t**3, 'g^')
plt.show()
```

##### linewidth arg for thickness of line

```python
plt.plot(x, y, linewidth=10.0)
```

##### setp() can be used to set line properties and works well with line objects

```python
lines = plt.plot(x, y)
plt.setp(lines, color='r', linewidth=2.0)
```

### plotting with keyword strings

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### plotting with cateogrical values

##### figsize = (height, width)

##### plt.subplot(131) means 1 x 3 grid for first graph

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### working with multiple figures and axes

##### subplot(num_rows, num_columns, plot_number)
##### plot_number range is from 1 to num_rows * num_columns 

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
plt.figure(1)                
plt.subplot(211)             
plt.plot([1, 2, 3])
plt.subplot(212)             
plt.plot([4, 5, 6])


plt.figure(2)                
plt.plot([4, 5, 6])
plt.title('not so easy fig 2')

plt.figure(1)                
plt.subplot(211)             
plt.title('Easy as 1, 2, 3') 
```

### working with text
##### The text() command can be used to add text in an arbitrary location
##### the xlabel(), ylabel() and title() are used to add text in the indicated locations 

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
